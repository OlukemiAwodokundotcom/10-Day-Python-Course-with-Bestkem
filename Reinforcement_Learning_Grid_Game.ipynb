# Create Environment

```python
import numpy as np
```

```python
grid_size = 3
```

```python
# (0,0) (0,1) (0,2)
# (1,0) (1,1) (1,2)
# (2,0) (2,1) (2,2)
```

```python
goal = (2,2)
```

```python
actions = ['up', 'down', 'left', 'right']
```

# Defining The Movement

```python
def move(state,action):
    x,y = state

    if action == 'up':
        x = max(0, x - 1)
    elif action == 'down':
        x = min(grid_size - 1, x+1)
    elif action == 'left':
        y = max(0,y-1)
    elif action == 'right':
        y = min(grid_size - 1, y+1)
    return(x,y)
```

# The Reward

```python
def get_reward(state):

    if state == goal:
        return 10

    else:
        return -1
```

# Training Loop

```python
state = (0,0)

for step in range(20):
    action = np.random.choice(actions)
    new_state = move(state,action)
    reward = get_reward(new_state)

    print(f"Step {step}: {state} -> {action} -> {new_state}, Reward: {reward}")

    state = new_state

    if state == goal:
          print("Goal reached")
          break

```
